```python
import json
```

```python
recipes_raw_path = "./data/recipes_raw.json"
ingredients      = []

# Go through the raw JSON recipes and pull out the ingredients into a huge array.
with open(recipes_raw_path) as f:
    data = json.load(f)
    for key in data:
        if "ingredients" in data[key]:
            for v in data[key]["ingredients"]:
                ingredients.append(v)
                
print(f"We have {len(ingredients)} individual ingredients\n")
```

```python
from parse_ingredients import parse_ingredient

output_path = "./data/training.json"
dataset     = []

n_training_samples  = 100000
max_sentence_length = 20

n = 0
for ingredient in ingredients:
    if n >= n_training_samples:
        break
    
    # Skip strings that are crazy long.
    if len(ingredient.split(' ')) > max_sentence_length:
        continue

    parsed = None
    try:
        parsed = parse_ingredient(ingredient)
    except:
        continue
    
    # Probably don't need 0 quantity homies for now. e.g "Dash of hot sauce"
    if parsed.quantity <= 0:
        continue
        
    if parsed.unit is None or parsed.unit == '':
        continue
    
    dataset.append({
        "x": ingredient,
        "y": { 
            "unit": parsed.unit,
            "quantity": parsed.quantity,
        }
    })
    
    n += 1
    
with open(output_path, 'w') as f:
    json.dump(dataset, f)

print(f"Done building dataset (sz: {len(dataset)}) to {output_path}")
```
